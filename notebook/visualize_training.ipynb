```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings

warnings.filterwarnings("ignore")
sns.set_style("darkgrid")
```

# Functions

```python
def plot_training(df):
    min_loss = np.argmax(df["validation_acc"])
    min_epoch = df["epoch"][min_loss]

    print(f"**\nBest model:\n{df.loc[min_epoch - 1, :]}\n**\n")

    fig, (ax1, ax2) = plt.subplots(2, figsize=(14, 12), sharex=True, sharey=False)
    sns.lineplot(data=df, x="epoch", y="validation_acc", label="validation", ax=ax1)
    sns.lineplot(data=df, x="epoch", y="train_acc", label="train", ax=ax1)
    sns.lineplot(data=df, x="epoch", y="test_acc", label="test", ax=ax1)
    ax1.set_ylabel("Validation")
    ax1.set_title("Train vs Validation vs Test Unweighted Accuracy")
    ax1.axvline(min_epoch, 0, color="red", linestyle="dotted")

    sns.lineplot(data=df, x="epoch", y="validation_loss", label="validation", ax=ax2)
    sns.lineplot(data=df, x="epoch", y="train_loss", label="train", ax=ax2)
    sns.lineplot(data=df, x="epoch", y="test_loss", label="test", ax=ax2)

    ax2.set_ylabel("Loss")
    ax2.set_title("Train vs Validation vs Test Loss")
    ax2.axvline(min_epoch, 0, color="red", linestyle="dotted")
    ax2.set_ylim(0, 3)
    plt.show()
```

# Analysis

```python
logs_path = f"../logs"
fold = 1

df = pd.read_csv(f"{logs_path}/fold_{fold}.csv", sep=",")
plot_training(df)
```
